```python
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
from sklearn.preprocessing import LabelEncoder

dataset = pd.read_csv('/content/new_yubi.csv')

label_encoder = LabelEncoder()

for column in dataset.columns:
    if dataset[column].dtype == 'object':
        dataset[column] = label_encoder.fit_transform(dataset[column])


threshold = 0.1


selector = VarianceThreshold(threshold)


selector.fit(dataset)


selected_feature_indices = selector.get_support(indices=True)


selected_features = dataset.columns[selected_feature_indices]


print(selected_features)
```

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
import numpy as np


df = pd.read_csv("/content/new_yubi.csv")


vectorizers = []


results = []

for column in df.columns:
    if df[column].dtype == 'object':

        df[column].fillna('', inplace=True)


        vectorizer = TfidfVectorizer(stop_words='english')


        X_tfidf = vectorizer.fit_transform(df[column])


        scaler = StandardScaler(with_mean=False)
        X_tfidf_scaled = scaler.fit_transform(X_tfidf)

        model = IsolationForest(contamination=0.05, random_state=42)
        anomaly_scores = model.fit_predict(X_tfidf_scaled)

        vectorizers.append(vectorizer)
        results.append({
            'column': column,
            'anomaly_scores': anomaly_scores,
        })


for result in results:
    print(f"Column: {result['column']}")
    print(f"Anomaly Scores: {result['anomaly_scores']}")
    print()
```

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler


df = pd.read_csv("/content/new_yubi.csv")


vectorizers = []


results = []

for column in df.columns:
    if df[column].dtype == 'object':

        df[column].fillna('', inplace=True)


        vectorizer = TfidfVectorizer(stop_words='english')


        X_tfidf = vectorizer.fit_transform(df[column])


        scaler = StandardScaler(with_mean=False)
        X_tfidf_scaled = scaler.fit_transform(X_tfidf)


        model = IsolationForest(contamination=0.05, random_state=42)
        anomaly_scores = model.fit_predict(X_tfidf_scaled)


        vectorizers.append(vectorizer)
        results.append({
            'column': column,
            'anomaly_scores': anomaly_scores,
            'anomaly_rows': df[anomaly_scores == -1],
        })

for result in results:
    print(f"Column: {result['column']}")
    print(f"Anomaly Scores: {result['anomaly_scores']}")
    print(f"Anomaly Rows:")
    print(result['anomaly_rows'])
    print()
```

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler


df = pd.read_csv("/content/new_yubi.csv")


vectorizers = []


results = []

for column in df.columns:
    if df[column].dtype == 'object':

        df[column].fillna('', inplace=True)


        vectorizer = TfidfVectorizer(stop_words='english')


        X_tfidf = vectorizer.fit_transform(df[column])


        scaler = StandardScaler(with_mean=False)
        X_tfidf_scaled = scaler.fit_transform(X_tfidf)


        model = IsolationForest(contamination=0.05, random_state=42)
        anomaly_scores = model.fit_predict(X_tfidf_scaled)


        vectorizers.append(vectorizer)
        results.append({
            'column': column,
            'anomaly_scores': anomaly_scores,
        })


for result in results:
    df[f"{result['column']}_anomaly_scores"] = result['anomaly_scores']


df.to_csv("/content/new_yubi_anomaly.csv", index=False)
```
